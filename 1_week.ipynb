```python
#1.
def clamp(x, lo, hi):
    lo2, hi2 = min(lo, hi), max(lo, hi)
    return max(lo2, min(x, hi2))

print(clamp(40, 8, 10))
```

```python
#2.
def safe_div(a, b, default=None):
    return a / b if b else default

print(safe_div(90, 0))
```

```python
#3.
def format_name(first: str, last: str, middle=None):
    return f'{last.capitalize()} {first[0].upper()}.{middle[0].upper()}.' if middle else  f'{last.capitalize()} {first[0].upper()}.'

print(format_name('vladislav', 'semenko', 'dmitrievich'))
```

```python
#4.
def sum_all(*args):
    return sum([e for e in args if type(e) in (int, float)]) if args else 0

print(sum_all())
```

```python
#5.
def only_ints(*args):
    return [x for x in args if type(x) is int]

print(only_ints(1, 3, 'd', True, 5, 'sdf'))
```

```python
#6.
def apply(func, numbers):
    return [func(n) for n in numbers]

def q(x):
    return x**2

print(apply(q, [e for e in range(4)]))
```

```python
#7.
def make_power(n):
    def power(x):
        return x**n
    return power

p = make_power(5)
print(p(2))
print(make_power(3)(5))
```

```python
#8.
def counter(start=0, step=1):
    count = 0
    def next_value():
        nonlocal count
        count += 1
        return start + count * step
    return next_value


value_plus = counter()

print(value_plus())
print(value_plus())
print(value_plus())
```

```python
#9.
def unique(seq):
    md = {}
    for e in seq:
        md[e] = md.get(e, 0) + 1
    return list(md.keys())


ml = [42, 'apple', 3.14, 'apple', True, (1, 2), 42, False, (1, 2), 'banana']
print(unique(ml))
```

```python
#10.
def fib(n, memo_d = {0: 0, 1: 1}):
    if n in memo_d:
        return memo_d[n]
    else:
        memo_d[n] = fib(n-1) + fib(n-2)
        return memo_d[n]

print(fib(7))
```

```python
#11.
def validate(callble_obj, *args, **kwargs):
    try:
        return (True, callble_obj(*args, **kwargs))
    except Exception as exp:
        return (False, type(exp))

print(validate(lambda x: 10/x, 0))
print(validate(max, (1, 2, 6, 8, 3, 5)))
```

```python
#12.
import functools, time


def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        value = func(*args, **kwargs)
        end = time.perf_counter()
        work_time = end - start
        print(f'{func.__name__}: {round(work_time, 4)} sec')
        return value
    return wrapper

@timer
def f(x):
    return x**(x**x)

res = f(5)

print(res)
```

```python
 #14.
import re

def normalize_spaces(text):
    return re.sub(r' +', ' ', text).strip()

txt1 = " Python   regex   иногда  выглядят   странно но   работают    отлично "
txt2 = "   Python   иногда   ведет   себя   неожиданно"
txt3 = "данные   могут   содержать    лишние   пробелы   "
txt4 = "  обработка  текста   в  Python   бывает    полезной   "

print(normalize_spaces(txt1))
print(normalize_spaces(txt2))
print(normalize_spaces(txt3))
print(normalize_spaces(txt4))
```

```python
#15.
def chunked(seq, size):
    res = []
    if size == 0:
        return []
    for i in range(len(seq)):
        if i % size == 0:
            chunk = []
            chunk.append(seq[i])
            if size == 1:
                res.append(chunk)
        elif i % size == (size - 1) :
            chunk.append(seq[i])
            res.append(chunk)
        else:
            chunk.append(seq[i])
    if chunk not in res:
        res.append(chunk)
    
    return res

data = ["яблоко", 42, 3.14, True, "Python", -7, "текст", 0]

print(chunked(data, 5))
```

```python
#16.
def flatten(list_of_lists: list):
    res = []
    for elem in list_of_lists:
        res.extend(elem)
    return res
data = [
    [3, 7, 12],
    [5, 9],
    [14, 2, 8, 11],
    [6],
    [10, 4, 1]
]

print(flatten(data))
```

```python
#17.
def merge_dicts(a: dict, b: dict) -> dict:
    a1, b1 = a, b
    a1.update(b1)
    return a1

dict1 = {
    "name": "Alice",
    "age": 25,
    "city": "Berlin",
    "job": "Engineer"
}

dict2 = {
    "name": "Bob",
    "age": 30,
    "city": "Paris",
    "salary": 5000
}


print(merge_dicts(dict1, dict2))
```

```python
#18.
def invert_dict(d: dict):
    new_dict = {}
    for k_v in d.items():
        new_dict[k_v[1]] = k_v[0]
    return new_dict

data = {
    "user1": "admin",
    "user2": "editor",
    "user3": "admin",
    "user4": "viewer",
    "user5": "editor",
    "user6": "admin"
}

print(invert_dict(data))
```

```python
#19.
def parse_int(s: str, base=10):
    try:
        return int(s, base)
    except:
        return None

print(parse_int('46', 2))
```

```python
#20.
import functools
def retry(callable_obj, attempts=3):
    @functools.wraps
    def wrapper(*args, **kwargs):
        try:
            value = callable_obj(*args, **kwargs)
            return value
        except:
            while True:
                if attempts != 0:
                    value = callable_obj(*args, **kwargs)
                    attempts -= 1
                return None
    return wrapper
```
